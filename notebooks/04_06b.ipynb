### The simplest time series plot

```python
address = '../data/Superstore-Sales.csv'
df = pd.read_csv(address, index_col='Order Date', encoding='cp1252', parse_dates=True)

df.head()
```

```python
plt.show(df['Order Quantity'].plot())
```
